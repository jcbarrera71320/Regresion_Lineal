```python
import numpy as np 
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn import linear_model 
```

```python
df= pd.read_csv('data.csv')
df 
```

```python
df.plot.scatter(x="metro" , y="precio")
plt.show()
```

```python
regresion= linear_model.LinearRegression() 
```

```python
metros = df["metro"].values.reshape((-1,1))
```

```python
modelo= regresion.fit(metros,df["precio"])

print("Intersección (b)", modelo.intercept_)
#Se imprime la pendiente 
print("Pendiente (m)", modelo.coef_)
```

```python
entrada= [[5],[15],[20],[25]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
df.plot.scatter(x="metro" , y="precio")
plt.scatter(entrada,predicciones, color ='red')
plt.plot(entrada,predicciones,color = 'black' , label = 'Linea de Regresión')
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()
```
